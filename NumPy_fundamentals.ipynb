```python
import numpy as np
```

```python
x=np.array(6)  #scalar (0-D tensor)
print("x:", x)
print("x ndim:", x.ndim)
print("x shape:", x.shape)
print("x size:", x.size)
print("x dtype:", x.dtype)
```

```python
x=np.array([1,2,3])  #Vector (1-D Tensor)
print("x:", x)
print("x ndim:", x.ndim)
print("x shape:", x.shape)   # 3 elements in ONE dimension. This is not a column or row yet (important!)
print("x size:", x.size)
print("x dtype:", x.dtype)
```

```python
x=np.array([[2,3],[4,5]])    # Matrix(2-D Tensor)
print("x:\n", x)
print("x ndim:", x.ndim)
print("x shape:", x.shape)    #2 rows Ã— 2 columns
print("x size:", x.size)
print("x dtype:", x.dtype)
```

```python
x=np.array([[[1,2],[3,5]],[[6,7],[8,9]]])
print("x:\n", x)
print("x ndim:", x.ndim)
print("x shape:", x.shape)   #2 blocks, 2 rows, 2 columns
print("x size:", x.size)
print("x dtype:", x.dtype)
```

```python
#Tensor Creation Functions (NumPy)
x=np.zeros((2,3))    #np.zeros()
print(x)
print("shape:", x.shape)
```

```python
x=np.ones((3,2))    #np.ones()
print(x)
print("shape:", x.shape)
```

```python
x=np.eye(3)    #np.eye()
print(x)
print("shape:", x.shape)
```

```python
x=np.random.random((2,3))    #np.random.random()
print(x)
print("shape:", x.shape)
```

```python
# Indexing & Slicing (Data Selection in ML)
x = np.array([10, 20, 30])   #Basic Indexing (1D Array)
print("x:", x)
print("x[0]:", x[0])

x[0] = 0
print("after change:", x)
```
